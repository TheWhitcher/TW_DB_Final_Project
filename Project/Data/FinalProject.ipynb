```python
#Import basic data tools
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Create dataframes from the different csv
co2 = 'annual-co2-emissions-per-country.csv'
gdpCapita = 'gdp-per-capita.csv'
methane = 'methane-emissions.csv'
nitrous = 'nitrous-oxide-emissions.csv'
popStat = 'population-and-demography.csv'
ghg = 'total-ghg-emissions.csv'

dfCo2 = pd.read_csv(co2)
dfGdp = pd.read_csv(gdpCapita)
dfMet = pd.read_csv(methane)
dfNit = pd.read_csv(nitrous)
dfPop = pd.read_csv(popStat)
dfGhg = pd.read_csv(ghg)
```

```python
#Make the first graph.
#Anual CO2 emissions per country.

#Select the countries to be displayed.
#Palce holder
countries = ["Canada", "United States", "United Kingdom", "China", "Japan"]

#Plot the dataframe with the seleced entities.
dfCo2_plot = dfCo2[(dfCo2["Entity"].isin(countries))]

#Select the style of the graph.
plt.style.use("ggplot")

fig , axs = plt.subplots() 

sns.lineplot(
    data = dfCo2_plot,
    x="Year",
    y="Annual COâ‚‚ emissions",
    hue="Entity",
    ax=axs,
)
plt.title("Annual COâ‚‚ emissions per Countries")
plt.savefig("annual COâ‚‚ emissions per Countries")
```
